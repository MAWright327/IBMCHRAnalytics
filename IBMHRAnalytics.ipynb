```python
#Data originally downloaded from: HR Analytics - https://www.kaggle.com/ludobenistant/hr-analytics/downloads/human-resources-analytics.zip

#import dependencies and import 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

filepath = '~/HR_comma_sep.csv'
df = pd.read_csv(filepath)

df[0:10]

```

```python
len(df.columns)
```

```python
df['salary'].unique()
```

```python
#OnHotEncode categorical variables.
from sklearn.preprocessing import OneHotEncoder

labels = df.left

data_enc = pd.get_dummies(df)
columns = [col for col in data_enc.columns if col != 'left']

data_enc.shape

```

```python
#View DataFrame with encoded data.
data_enc[0:10]
```

```python
#Separate columns to be used for analysis.
data = data_enc[[col for col in data_enc.columns if col != 'left']]

data[0:10]
```

```python
#Separate dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.33, random_state=42)
```

```python
#Use PCA for dimensionality reduction to obtain Top 20 Components and view explained variance of each variable.
from sklearn.decomposition import PCA

pca = PCA(n_components=20)
pca.fit(X_train)

exp_var = pca.explained_variance_ratio_
exp_var = [round(x, 5) for x in exp_var]

df_expVar = pd.DataFrame(exp_var, columns)
df_expVar.columns = ['Explained Var']

red_cols = list(df_expVar.index[0:20])
red_cols
df_expVar
```

```python
#Obtain DataFrame with columns from top 20 components from PCA.
data_red = data[[col for col in data_enc.columns if col in red_cols]]

data_red[0:10]
```

```python
#Obtain new train and test datasets from the reduced dataset.
X_train, X_test, y_train, y_test = train_test_split(data_red, labels, test_size=0.33, random_state=42)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

clf_lr = LogisticRegression()
clf_lr.fit(X_train, y_train)
predicted_lr = clf_lr.predict(X_test)

print(classification_report(y_test, predicted_lr, target_names=['No', 'Yes']))
```

```python
clf_lr.score(X_test,y_test)
```

```python
#SVM
from sklearn.svm import SVC

clf_svc = SVC()
clf_svc.fit(X_train, y_train)
predicted_svc = clf_svc.predict(X_test)
print(classification_report(y_test, predicted_svc, target_names=['No', 'Yes']))
```

```python
clf_svc.score(X_test,y_test)
```

```python
#Decision Tree
from sklearn import tree

clf_tree = tree.DecisionTreeClassifier()
clf_tree.fit(X_train, y_train)
predicted_tree = clf_tree.predict(X_test)
print(classification_report(y_test, predicted_tree, target_names=['No', 'Yes']))
```

```python
#View most important features from Decision Tree Analysis.
features = clf_tree.feature_importances_
df_tree_features = pd.DataFrame(features, columns)
df_tree_features.columns = ['Feature Importance']
df_tree_features
```

```python
clf_tree.score(X_test,y_test)
```

```python
#Ensemble using Decision Tree and AdaBoost
from sklearn.ensemble import AdaBoostClassifier

n_estimators = 100
learning_rate = 1

adaBoost = AdaBoostClassifier(base_estimator=clf_tree, n_estimators=n_estimators, 
                              learning_rate=learning_rate)
adaBoost.fit(X_train, y_train)
adaBoost_pred = adaBoost.predict(X_test)

print(classification_report(y_test, adaBoost_pred, target_names=['No', 'Yes']))
```

```python
adaBoost.score(X_test,y_test)
```

```python
#Ensemble using Decision Tree and Bagging
from sklearn.ensemble import BaggingClassifier

n_estimators = 100
learning_rate = 1

bagging = BaggingClassifier(base_estimator=clf_tree, n_estimators=n_estimators, 
                              bootstrap=True)
bagging.fit(X_train, y_train)
bagging_pred = bagging.predict(X_test)

print(classification_report(y_test, bagging_pred, target_names=['No', 'Yes']))
```

```python
bagging.score(X_test,y_test)
```
